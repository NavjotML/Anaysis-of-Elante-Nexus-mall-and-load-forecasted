```python
# 3. SOLAR VS GRID CONSUMPTION (Grouped Bar Chart)
plt.figure(figsize=(12, 6))
solar_cols = ['Mall Solar', 'OB Solar', 'Hotel  Solar']
total_solar = df[solar_cols].sum(axis=1)
grid_consumption = df['Consumption'] - total_solar

x = range(len(df))
width = 0.35

plt.bar([i - width/2 for i in x], grid_consumption, width, label='Grid', alpha=0.8, color='steelblue')
plt.bar([i + width/2 for i in x], total_solar, width, label='Solar', alpha=0.8, color='gold')

plt.xticks(x, df['Months'], rotation=45, ha='right')
plt.title('Solar vs Grid Consumption', fontsize=14, fontweight='bold')
plt.xlabel('')
plt.ylabel('kWh')
plt.legend()
plt.ylim(0, None)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('3_solar_vs_grid.png', dpi=300)
plt.show()
```

```python
# 3. SOLAR VS GRID CONSUMPTION (Pie Chart - Annual Total)
plt.figure(figsize=(10, 8))

solar_cols = ['Mall Solar', 'OB Solar', 'Hotel  Solar']
total_solar_year = df[solar_cols].sum().sum()  # Total solar for the year
total_consumption_year = df['Consumption'].sum()  # Total consumption for the year
grid_consumption_year = total_consumption_year - total_solar_year

# Data for pie chart
sizes = [grid_consumption_year, total_solar_year]
labels = ['Grid Consumption', 'Solar Generation']
colors = ['steelblue', 'gold']
explode = (0, 0.1)  # Explode the solar slice slightly

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, 
        colors=colors, explode=explode, shadow=True, textprops={'fontsize': 12})
plt.title('Annual Energy Source Distribution (Grid vs Solar)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('3_solar_vs_grid_pie.png', dpi=300)
plt.show()

# Print actual values
print(f"Total Grid Consumption: {grid_consumption_year:,.0f} kWh")
print(f"Total Solar Generation: {total_solar_year:,.0f} kWh")
print(f"Solar Percentage: {(total_solar_year/total_consumption_year)*100:.2f}%")
```

```python
# 4. DG (DIESEL GENERATOR) USAGE PATTERN (Stacked Bar)
plt.figure(figsize=(12, 6))
dg_cols = ['Mall DG', 'Tenant  DG', 'HVAC DG', 'CA DG', 'OB DG ', 'Hotel DG']

df[dg_cols].plot(kind='bar', stacked=True, figsize=(12, 6), width=0.8)

plt.title('Diesel Generator Usage by Facility', fontsize=14, fontweight='bold')
plt.xlabel('')
plt.ylabel('DG Consumption (kWh)')
plt.xticks(range(12), df['Months'], rotation=45, ha='right')
plt.legend(title='Facility DG', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.ylim(0, None)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('4_dg_usage_pattern.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 5. HVAC CONSUMPTION ANALYSIS (Line + Bar Combo)
fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar chart for HVAC on left axis
ax1.bar(df['Months'], df['HVAC '], alpha=0.6, label='HVAC Consumption', color='steelblue', width=0.6)
ax1.set_xlabel('')
ax1.set_ylabel('HVAC Consumption (kWh)', color='steelblue', fontsize=11)
ax1.tick_params(axis='y', labelcolor='steelblue')
plt.xticks(rotation=45, ha='right')
ax1.set_ylim(0, None)
ax1.grid(True, alpha=0.3, axis='y')

# Line chart for Total Consumption on right axis
ax2 = ax1.twinx()
ax2.plot(df['Months'], df['Consumption'], marker='o', color='red', linewidth=2, 
         markersize=8, label='Total Consumption')
ax2.set_ylabel('Total Consumption (kWh)', color='red', fontsize=11)
ax2.tick_params(axis='y', labelcolor='red')
ax2.set_ylim(0, None)

plt.title('HVAC Consumption vs Total Consumption', fontsize=14, fontweight='bold')

# Combine legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

fig.tight_layout()
plt.savefig('5_hvac_analysis.png', dpi=300)
plt.show()
```

```python
# 6. FACILITY CONSUMPTION COMPARISON (Box Plot)
plt.figure(figsize=(12, 6))
facilities = ['Mall ', 'Tenant ', 'HVAC ', 'CA  Lighting', 'OB', 'Hotel ']

facility_data = df[facilities].melt(var_name='Facility', value_name='Consumption')
sns.boxplot(data=facility_data, x='Facility', y='Consumption', palette='Set2')

plt.title('Consumption Distribution by Facility Type', fontsize=14, fontweight='bold')
plt.xlabel('')
plt.ylabel('Consumption (kWh)')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, None)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('6_facility_comparison_boxplot.png', dpi=300)
plt.show()
```

```python
# 7. PERCENTAGE CONTRIBUTION PIE CHART (Annual Average)
plt.figure(figsize=(10, 8))
facilities = ['Mall ', 'Tenant ', 'HVAC ', 'CA  Lighting', 'OB', 'Hotel ']

annual_totals = df[facilities].sum()
colors = sns.color_palette('husl', len(facilities))

plt.pie(annual_totals, labels=facilities, autopct='%1.1f%%', startangle=90, 
        colors=colors, textprops={'fontsize': 11})
plt.title('Annual Consumption Share by Facility', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('7_facility_pie_chart.png', dpi=300)
plt.show()

# Print actual values
print("\nAnnual Consumption by Facility:")
for facility in facilities:
    total = df[facility].sum()
    percentage = (total / df[facilities].sum().sum()) * 100
    print(f"{facility}: {total:,.0f} kWh ({percentage:.1f}%)")
```

```python
# 8. RENEWABLE ENERGY CONTRIBUTION (Stacked Percentage)
plt.figure(figsize=(12, 6))

solar_cols = ['Mall Solar', 'OB Solar', 'Hotel  Solar']
renewable = df[solar_cols].sum(axis=1)
non_renewable = df['Consumption'] - renewable

percentage_df = pd.DataFrame({
    'Solar': (renewable / df['Consumption'] * 100),
    'Grid': (non_renewable / df['Consumption'] * 100)
})

percentage_df.plot(kind='bar', stacked=True, color=['gold', 'gray'], 
                   alpha=0.8, figsize=(12, 6), width=0.8)

plt.title('Renewable vs Non-Renewable Energy Percentage', fontsize=14, fontweight='bold')
plt.xlabel('')
plt.ylabel('Percentage (%)')
plt.xticks(range(12), df['Months'], rotation=45, ha='right')
plt.legend(title='Energy Source', labels=['Solar (Renewable)', 'Grid (Non-Renewable)'])
plt.ylim(0, 100)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('8_renewable_percentage.png', dpi=300, bbox_inches='tight')
plt.show()

# Print monthly solar percentages
print("\nMonthly Solar Energy Percentage:")
for idx, month in enumerate(df['Months']):
    solar_pct = (renewable.iloc[idx] / df['Consumption'].iloc[idx]) * 100
    print(f"{month}: {solar_pct:.2f}%")
```

```python
# 9. HEATMAP OF FACILITY CONSUMPTION
plt.figure(figsize=(14, 8))
facilities = ['Mall ', 'Tenant ', 'HVAC ', 'CA  Lighting', 'OB', 'Hotel ']

heatmap_data = df[facilities].T
sns.heatmap(heatmap_data, annot=True, fmt='.0f', cmap='YlOrRd', 
            xticklabels=df['Months'], yticklabels=facilities, 
            cbar_kws={'label': 'Consumption (kWh)'}, linewidths=0.5)

plt.title('Monthly Consumption Heatmap by Facility', fontsize=14, fontweight='bold')
plt.xlabel('')
plt.ylabel('Facility Type')
plt.tight_layout()
plt.savefig('9_consumption_heatmap.png', dpi=300)
plt.show()
```

```python
# 10. SEASONAL ANALYSIS (Grouped Bar Chart)
plt.figure(figsize=(10, 6))

# Define seasons
seasons = {
    'Winter': ['January', 'February', 'December'],
    'Spring': ['March', 'April', 'May'],
    'Summer': ['June', 'July', 'August'],
    'Fall': ['September', 'October', 'November']
}

seasonal_data = []
for season, months in seasons.items():
    season_df = df[df['Months'].isin(months)]
    solar_cols = ['Mall Solar', 'OB Solar', 'Hotel  Solar']
    dg_cols = ['Mall DG', 'Tenant  DG', 'HVAC DG', 'CA DG', 'OB DG ', 'Hotel DG']
    
    seasonal_data.append({
        'Season': season,
        'Avg Consumption': season_df['Consumption'].mean(),
        'Avg Solar': season_df[solar_cols].sum(axis=1).mean(),
        'Avg DG': season_df[dg_cols].sum(axis=1).mean()
    })

seasonal_df = pd.DataFrame(seasonal_data)
seasonal_df.set_index('Season')[['Avg Consumption', 'Avg Solar', 'Avg DG']].plot(
    kind='bar', figsize=(10, 6), width=0.7)

plt.title('Seasonal Energy Consumption Analysis', fontsize=14, fontweight='bold')
plt.xlabel('')
plt.ylabel('Average Consumption (kWh)')
plt.xticks(rotation=45, ha='right')
plt.legend(['Total Consumption', 'Solar Generation', 'DG Usage'])
plt.ylim(0, None)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('10_seasonal_analysis.png', dpi=300)
plt.show()

# Print seasonal insights
print("\nSeasonal Analysis:")
for _, row in seasonal_df.iterrows():
    print(f"\n{row['Season']}:")
    print(f"  Avg Consumption: {row['Avg Consumption']:,.0f} kWh")
    print(f"  Avg Solar: {row['Avg Solar']:,.0f} kWh")
    print(f"  Avg DG: {row['Avg DG']:,.0f} kWh")
```

```python
# COMPARISON GRAPH 5: Solar Percentage Comparison
plt.figure(figsize=(12, 6))

# Calculate solar percentage for each year
solar_2023 = df[['Mall Solar', 'OB Solar', 'Hotel  Solar']].sum(axis=1)
solar_pct_2023 = (solar_2023 / df['Consumption'] * 100)

solar_2024 = df_2024['Total_Solar'][:12]
solar_pct_2024 = (solar_2024 / df_2024['Consumption'][:12] * 100)

x = range(12)
width = 0.35

plt.bar([i - width/2 for i in x], solar_pct_2023, width, label='2023', alpha=0.8, color='gold')
plt.bar([i + width/2 for i in x], solar_pct_2024, width, label='2024', alpha=0.8, color='darkorange')

plt.xticks(x, df['Months'], rotation=45, ha='right')
plt.title('Solar Energy Percentage Comparison: 2023 vs 2024', fontsize=14, fontweight='bold')
plt.xlabel('')
plt.ylabel('Solar Percentage (%)')
plt.legend(fontsize=12)
plt.ylim(0, None)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('comparison_5_solar_percentage.png', dpi=300)
plt.show()

# Print comparison
print("\nSolar Percentage Comparison:")
print(f"2023 Average Solar %: {solar_pct_2023.mean():.2f}%")
print(f"2024 Average Solar %: {solar_pct_2024.mean():.2f}%")
print(f"Improvement: {(solar_pct_2024.mean() - solar_pct_2023.mean()):.2f} percentage points")

print("\n✅ Comparison Graph 5 created!")
```

```python
# COMPARISON GRAPH 10: Annual Summary Dashboard
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Annual Energy Performance Dashboard: 2023 vs 2024', fontsize=16, fontweight='bold')

# Metric 1: Total Consumption
ax1 = axes[0, 0]
total_2023 = df['Consumption'].sum()
total_2024 = df_2024['Consumption'][:12].sum()
ax1.bar(['2023', '2024'], [total_2023/1e6, total_2024/1e6], color=['steelblue', 'orangered'], alpha=0.8)
ax1.set_title('Total Annual Consumption', fontweight='bold')
ax1.set_ylabel('Million kWh')
ax1.grid(True, alpha=0.3, axis='y')
for i, (year, val) in enumerate(zip(['2023', '2024'], [total_2023/1e6, total_2024/1e6])):
    ax1.text(i, val, f'{val:.2f}M', ha='center', va='bottom', fontweight='bold')

# Metric 2: Total Solar Generation
ax2 = axes[0, 1]
solar_total_2023 = df[['Mall Solar', 'OB Solar', 'Hotel  Solar']].sum().sum()
solar_total_2024 = df_2024['Total_Solar'][:12].sum()
ax2.bar(['2023', '2024'], [solar_total_2023/1e3, solar_total_2024/1e3], color=['gold', 'orange'], alpha=0.8)
ax2.set_title('Total Solar Generation', fontweight='bold')
ax2.set_ylabel('Thousand kWh')
ax2.grid(True, alpha=0.3, axis='y')
for i, (year, val) in enumerate(zip(['2023', '2024'], [solar_total_2023/1e3, solar_total_2024/1e3])):
    ax2.text(i, val, f'{val:.1f}K', ha='center', va='bottom', fontweight='bold')

# Metric 3: Total DG Usage
ax3 = axes[1, 0]
dg_2023 = df[['Mall DG', 'Tenant  DG', 'HVAC DG', 'CA DG', 'OB DG ', 'Hotel DG']].sum().sum()
dg_2024 = df_2024[['Hotel_DG', 'HVAC_Tenants_DG', 'HVAC_CA_DG', 'Lighting_Tenants_DG', 
                   'Lighting_CA_DG', 'Office_Tenants_DG', 'Office_CA_DG']][:12].sum().sum()
ax3.bar(['2023', '2024'], [dg_2023/1e3, dg_2024/1e3], color=['crimson', 'darkred'], alpha=0.8)
ax3.set_title('Total DG Usage', fontweight='bold')
ax3.set_ylabel('Thousand kWh')
ax3.grid(True, alpha=0.3, axis='y')
for i, (year, val) in enumerate(zip(['2023', '2024'], [dg_2023/1e3, dg_2024/1e3])):
    ax3.text(i, val, f'{val:.1f}K', ha='center', va='bottom', fontweight='bold')

# Metric 4: Average Solar Percentage
ax4 = axes[1, 1]
solar_pct_2023 = (solar_total_2023 / total_2023 * 100)
solar_pct_2024 = (solar_total_2024 / total_2024 * 100)
ax4.bar(['2023', '2024'], [solar_pct_2023, solar_pct_2024], color=['gold', 'orange'], alpha=0.8)
ax4.set_title('Solar Contribution %', fontweight='bold')
ax4.set_ylabel('Percentage (%)')
ax4.grid(True, alpha=0.3, axis='y')
for i, (year, val) in enumerate(zip(['2023', '2024'], [solar_pct_2023, solar_pct_2024])):
    ax4.text(i, val, f'{val:.2f}%', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.savefig('comparison_10_annual_dashboard.png', dpi=300)
plt.show()

# Print summary
print("\n" + "="*60)
print("ANNUAL ENERGY PERFORMANCE SUMMARY")
print("="*60)
print(f"\nTotal Consumption:")
print(f"  2023: {total_2023:,.0f} kWh")
print(f"  2024: {total_2024:,.0f} kWh")
print(f"  Change: {((total_2024-total_2023)/total_2023*100):+.2f}%")

print(f"\nTotal Solar Generation:")
print(f"  2023: {solar_total_2023:,.0f} kWh")
print(f"  2024: {solar_total_2024:,.0f} kWh")
print(f"  Change: {((solar_total_2024-solar_total_2023)/solar_total_2023*100):+.2f}%")

print(f"\nTotal DG Usage:")
print(f"  2023: {dg_2023:,.0f} kWh")
print(f"  2024: {dg_2024:,.0f} kWh")
print(f"  Change: {((dg_2024-dg_2023)/dg_2023*100):+.2f}%")

print(f"\nSolar Contribution:")
print(f"  2023: {solar_pct_2023:.2f}%")
print(f"  2024: {solar_pct_2024:.2f}%")
print(f"  Improvement: {(solar_pct_2024-solar_pct_2023):+.2f} percentage points")
print("="*60)

print("\n✅ Comparison Graph 10 created! All graphs complete! 🎉")
```
